# **Data Understanding**

```python
# Import necessary libraries:
import numpy as np                   # numpy for numerical operations
import pandas as pd                  # pandas for data manipulation and analysis
import matplotlib.pyplot as plt      # matplotlib.pyplot for data visualization
import seaborn as sns                # seaborn library for advanced data visualization
```

```python
# Load the dataset from a CSV file into a pandas DataFrame
df = pd.read_csv('data.csv')
```

```python
# Display the first few rows of the dataset to get an overview
df.head()
```

```python
# Get the dimensions of the DataFrame (number of rows and columns)
df.shape
```

```python
# Display concise summary of the DataFrame, including: column names, data types, non-null counts, and memory usage
df.info()
```

```python
# Count the total number of duplicate rows in the DataFrame
df.duplicated().sum()
```

```python
# Count the number of missing (null) values in each column of the DataFrame
df.isnull().sum()
```

```python
# Drop the 'Unnamed: 0' column from the DataFrame permanently
# columns=['Unnamed: 0'] → specifies the name of the column to be dropped.
# inplace=True → specifies that delete the column permanently.

df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
# Display the first few rows again to confirm that the 'Unnamed: 0' column has been removed
df.head()
```

```python
# Remove 'GB' so RAM is just a number (needed for models)
df['Ram'] = df['Ram'].str.replace('GB', '')

# Remove 'kg' so Weight is just a number (needed for models)
df['Weight'] = df['Weight'].str.replace('kg', '')

# Removing 'GB' and 'kg' makes the RAM and Weight columns pure numbers,
# so they can be used directly fed into machine learning models without errors.
```

```python
# Show the first few rows to check that 'GB' and 'kg' were removed correctly
df.head()
```

```python
# Convert 'Ram' to integer and 'Weight' to float so they are proper numeric types for modeling
df['Ram'] = df['Ram'].astype('int32')
df['Weight'] = df['Weight'].astype('float32')
```

```python
# Check the updated data types to confirm that 'Ram' is int and 'Weight' is float
df.info()
```

```python
# Plot the number of laptops from each company as a bar chart
df['Company'].value_counts().plot(kind='bar')
```

```python
# this shows the average laptop price for each company.
sns.barplot(x=df['Company'],y=df['Price'])
plt.xticks(rotation='vertical')
plt.show()
```

```python
# Plot the count of each "laptop type" as a bar chart
# This helps us to see which laptop types are most common in our dataset.
df['TypeName'].value_counts().plot(kind='bar')
```
